### 以Google Map API抓取各縣市半徑750公尺內的店家資訊

```python
import googlemaps
import pandas as pd
import time
gmaps = googlemaps.Client(key='Your API key') 
```

```python
#創造一個 list 放入台北市各區資料
with open("台北市里名.txt", "r", encoding="utf-8") as f:
    cities = f.read().splitlines()
    print(cities)
```

```python
#檢視gmaps.geocode回傳的格式
# geocode_result = gmaps.geocode("臺北市")
# geocode_result
```

```python
#先創一個名為 ids 的 list，寫一個迴圈利用 gmaps.geocode 函式去獲得每一個縣市GPS位址
#搜尋該縣市位址中心半徑 750 公尺符合咖啡廳的店家 id，並將這些 id append 到先前創好的 ids 變數中
ids = []
for city in cities:
    results = []
    # Geocoding an address
    geocode_result = gmaps.geocode(city)
    loc = geocode_result[0]['geometry']['location']
    query_result = gmaps.places_nearby(keyword="restaurant", location=loc, radius = 750) #設定商店類別 ex:restaurant,咖啡廳,麵包店,酒吧
    results.extend(query_result['results']) 
    while query_result.get('next_page_token'):
        time.sleep(2)
        query_result = gmaps.places_nearby(page_token = query_result['next_page_token'])
        results.extend(query_result['results'])    
    print("找到以"+city+"為中心半徑750公尺的店家數量(google map api上限提供60間): "+str(len(results)))
    for place in results:
        ids.append(place['place_id'])
```

```python
stores_info = []
# 利用 list,set 方式去除重複的店家資訊(id)
ids = list(set(ids)) 
# 利用 gmaps.place 函式取得店家資訊
for id in ids:
    stores_info.append(gmaps.place(place_id=id, language='zh-TW')['result'])
```

```python
#將資料轉成 DataFrame 格式
output = pd.DataFrame.from_dict(stores_info)
```

```python
# 抓出經度與緯度
output['lat'] = output['geometry'].map(lambda x: x['location']['lat'])
output['lng'] = output['geometry'].map(lambda x: x['location']['lng'])
```

```python
#刪除不必要的欄位
output = output.drop(columns=['address_components','adr_address','geometry','photos','scope','international_phone_number','icon','utc_offset','vicinity','reference'])
```
